**Polynomial Kernel**\
The C value (upper bound of alpha parameter) had to be changed to allow for some datapoints inside the margins for the model to work properly.

```python
polynomialM = PolynomialSV(inputs, targets, classA, classB, C=None)
polynomialM.train(p=3, alpha_threshold=-5)
polynomialM.plot()
```
